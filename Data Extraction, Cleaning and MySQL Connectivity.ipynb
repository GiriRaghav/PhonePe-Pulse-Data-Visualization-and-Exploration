# Aggregated Transaction Data

```python
#For aggregated transaction data 

#Path to get the data 
path = "C:/Users/GIRI RAGHAV/Phonepe Pulse Project/pulse/data/aggregated/transaction/country/india/state/"
Agg_trans_data=os.listdir(path)
Agg_trans_data
```

```python
#To extract the data and create a DataFrame

clm1={'State':[],'Year':[],'Quarter':[],'Transaction_type':[],'Transaction_count':[],'Transaction_amount':[]}

for i in Agg_trans_data:
    p_i = path+i+"/"
    Agg_yr = os.listdir(p_i)
    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list= os.listdir(p_j)        
        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            
            for z in D['data']['transactionData']:
                Name=z['name']
                count=z['paymentInstruments'][0]['count']
                amount=z['paymentInstruments'][0]['amount']
                clm1['Transaction_type'].append(Name)
                clm1['Transaction_count'].append(count)
                clm1['Transaction_amount'].append(amount)
                clm1['State'].append(i)
                clm1['Year'].append(j)
                clm1['Quarter'].append(int(k.strip('.json')))
```

```python
#Succesfully created a DataFrame
Agg_trans_data=pd.DataFrame(clm1)
Agg_trans_data
```

# Aggregated User Data

# Map Transaction Data

```python
#For map transaction data 

#Path to get the data 
path = "C:/Users/GIRI RAGHAV/Phonepe Pulse Project/pulse/data/map/transaction/hover/country/india/state/"
map_trans_data = os.listdir(path)
map_trans_data
```

```python
#To extract the data and create a DataFrame

clm3={'State':[], 'Year':[],'Quarter':[],'District': [],'Transaction_count':[],'Transaction_amount':[]}

for i in map_trans_data:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i) 

    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j) 

        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            H=json.load(Data)

            for z in H['data']['hoverDataList']:
                District = z['name']
                count = z["metric"][0]["count"]
                amount = z["metric"][0]["amount"]
                clm3["District"].append(District)
                clm3['Transaction_count'].append(count)
                clm3['Transaction_amount'].append(amount)
                clm3['State'].append(i)
                clm3['Year'].append(j)
                clm3['Quarter'].append(int(k.strip('.json')))
      
```

# Top Transaction Data

```python
#To extract the data and create a DataFrame

clm5={'State':[],'Year':[],'Quarter':[],'District':[],'Transaction_count':[],'Transaction_amount':[]}

for i in top_trans_data:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i) 
    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j) 

        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            T=json.load(Data)
            
            for z in T['data']['pincodes']:
                Name = z['entityName']
                count = z['metric']['count']
                amount = z['metric']['amount']
                clm5['District'].append(Name)
                clm5['Transaction_count'].append(count)
                clm5['Transaction_amount'].append(amount)
                clm5['State'].append(i)
                clm5['Year'].append(j)
                clm5['Quarter'].append(int(k.strip('.json')))
```

```python
import mysql.connector
import pandas as pd
```

```python
#Create a connection to the MySQL database
mydb = mysql.connector.connect(
    host="localhost",
    user="your_username",
    password="your_password",
    database="phonepe_pulse_database",
    auth_plugin='mysql_native_password')
```

```python
#Create cursor object
cursor = mydb.cursor()
```

```python
#Define SQL statements for creating tables
agg_trans_table = '''CREATE TABLE IF NOT EXISTS agg_trans_data (
                        id INT AUTO_INCREMENT PRIMARY KEY,
                        State VARCHAR(255),
                        Year INT,
                        Quarter INT,
                        Transaction_type VARCHAR(255),
                        Transaction_count INT,
                        Transaction_amount FLOAT
                    )'''

agg_user_table = '''CREATE TABLE IF NOT EXISTS agg_user (
                        id INT AUTO_INCREMENT PRIMARY KEY,
                        State VARCHAR(255),
                        Year INT,
                        Quarter INT,
                        Brands VARCHAR(50),
                        User_count INT,
                        User_percentage FLOAT
                    )'''

map_trans_table = '''CREATE TABLE IF NOT EXISTS map_trans_data (
                        id INT AUTO_INCREMENT PRIMARY KEY,
                        State VARCHAR(255),
                        Year INT,
                        Quarter INT,
                        District VARCHAR(255),
                        Transaction_count INT,
                        Transaction_amount FLOAT
                    )'''

map_user_table = '''CREATE TABLE IF NOT EXISTS map_user_data (
                        id INT AUTO_INCREMENT PRIMARY KEY,
                        State VARCHAR(255),
                        Year INT,
                        Quarter INT,
                        District VARCHAR(255),
                        Number_of_users INT
                    )'''

top_trans_table = '''CREATE TABLE IF NOT EXISTS top_trans_data (
                        id INT AUTO_INCREMENT PRIMARY KEY,
                        State VARCHAR(255),
                        Year INT,
                        Quarter INT,
                        District VARCHAR(255),
                        Transaction_count INT,
                        Transaction_amount FLOAT
                    )'''

top_user_table = '''CREATE TABLE IF NOT EXISTS top_user_data (
                        id INT AUTO_INCREMENT PRIMARY KEY,
                        State VARCHAR(255),
                        Year INT,
                        Quarter INT,
                        District VARCHAR(255),
                        Registered_users INT
                    )'''
```

```python
#Create tables
cursor.execute(agg_trans_table)
cursor.execute(agg_user_table)
cursor.execute(map_trans_table)
cursor.execute(map_user_table)
cursor.execute(top_trans_table)
cursor.execute(top_user_table)
```

```python
#Define SQL statements for inserting data from CSV files
agg_trans_insert = '''INSERT INTO agg_trans_data (
                            State, Year, Quarter, Transaction_type, Transaction_count, Transaction_amount
                        ) VALUES (%s, %s, %s, %s, %s, %s)'''

agg_user_insert = '''INSERT INTO agg_user (
                            State, Year, Quarter, Brands, User_count, User_percentage
                        ) VALUES (%s, %s, %s, %s, %s, %s)'''

map_trans_insert = '''INSERT INTO map_trans_data (
                            State, Year, Quarter, District, Transaction_count, Transaction_amount
                        ) VALUES (%s, %s, %s, %s, %s, %s)'''

map_user_insert = '''INSERT INTO map_user_data (
                            State, Year, Quarter, District, Number_of_users
                        ) VALUES (%s, %s, %s, %s, %s)'''

top_trans_insert = '''INSERT INTO top_trans_data (
                            State, Year, Quarter, District, Transaction_count, Transaction_amount
                        ) VALUES (%s, %s, %s, %s, %s, %s)'''

top_user_insert = '''INSERT INTO top_user_data (
                            State, Year, Quarter, District, Registered_users
                        ) VALUES (%s, %s, %s, %s, %s)'''
```

```python
#Load CSV files into DataFrames
agg_trans_df = pd.read_csv('C:/Users/GIRI RAGHAV/Phonepe Pulse Project/Agg_trans_data.csv')
agg_user_df = pd.read_csv('C:/Users/GIRI RAGHAV/Phonepe Pulse Project/Agg_user.csv')
map_trans_df = pd.read_csv('C:/Users/GIRI RAGHAV/Phonepe Pulse Project/map_trans_data.csv')
map_user_df = pd.read_csv('C:/Users/GIRI RAGHAV/Phonepe Pulse Project/map_user_data.csv')
top_trans_df = pd.read_csv('C:/Users/GIRI RAGHAV/Phonepe Pulse Project/top_trans_data.csv')
top_user_df = pd.read_csv('C:/Users/GIRI RAGHAV/Phonepe Pulse Project/top_user_data.csv')
```

```python
# Insert data into tables
for row in agg_trans_df.itertuples():
    cursor.execute(agg_trans_insert, (row.State, row.Year, row.Quarter, row.Transaction_type, row.Transaction_count, row.Transaction_amount))

for row in agg_user_df.itertuples():
    cursor.execute(agg_user_insert, (row.State, row.Year, row.Quarter, row.Brands, row.User_count, row.User_percentage))
    
for row in map_trans_df.itertuples():
    cursor.execute(map_trans_insert, (row.State, row.Year, row.Quarter, row.District, row.Transaction_count, row.Transaction_amount))

for row in map_user_df.itertuples():
    cursor.execute(map_user_insert, (row.State, row.Year, row.Quarter, row.District, row.Number_of_users))
                                     
for row in top_trans_df.itertuples():
    cursor.execute(top_trans_insert, (row.State, row.Year, row.Quarter, row.District, row.Transaction_count, row.Transaction_amount))
                                    
for row in top_user_df.itertuples():
    cursor.execute(top_user_insert, (row.State, row.Year, row.Quarter, row.District, row.Registered_users))
```

```python
#Commit changes and close cursor and database connection
mydb.commit()
cursor.close()
mydb.close()
```
